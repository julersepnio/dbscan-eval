### ðŸŒ² Robust Random Cut Forest
* Because RRCF is fundamentally a classification tool, our metrics must go beyond simple observations. RRCF produces a 'displacement' score for every point, but the value of the model lies in where we draw the line to label something an anomaly.

  * To be 'strict' in our measurement, we should focus on:
    * Recall: Are we catching all the actual anomalies?
    * Precision: When we flag an item, how often are we right?
    * The F1-Score: Finding the optimal balance between the two.

* Our baseline testing and the visualization above demonstrates high recall

### ðŸ”Ž Streamable DBSCAN

DBSCAN doesn't provide a direct anomaly 'score'; instead, it groups data based on density. To use it effectively for real-time monitoring, we must look beyond simple outlier detection and instead track cluster changes. This means monitoring how clusters drift, shrink, or grow over time, as these structural fluctuations often represent more significant system changes than isolated data anomalies.
