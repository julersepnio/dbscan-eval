## [üåê Web Traffic Dataset](https://www.kaggle.com/datasets/raminhuseyn/web-traffic-time-series-dataset)

### üí≠ Datataset Analysis

```python
import kagglehub
import pandas as pd
import matplotlib.pyplot as plt
from comprehensive_eda import comprehensive_eda

# Download latest version
path = kagglehub.dataset_download("raminhuseyn/web-traffic-time-series-dataset")
print("Path to dataset files:", path)

# Load Raw Data
web_traffic_df = pd.read_csv(path + '/web_traffic.csv')

# Run the comprehensive EDA given
comprehensive_eda(web_traffic_df)

# Plot the overall Timeseries Data
# Used the scatter plot to visualize how well the time-series data would be clustered
plt.figure(figsize=(15, 5))
plt.scatter(web_traffic_df['Timestamp'], web_traffic_df['TrafficCount'], marker='x')
plt.title('Web Traffic Time Series Dataset')
plt.xlabel(None)

# Zoom in the data
plt.figure(figsize=(15, 5))
plt.xticks(rotation=90)
plt.scatter(web_traffic_df['Timestamp'][0:100], web_traffic_df['TrafficCount'][0:100], marker='x')
plt.title('Web Traffic Time Series Dataset(Zoomed-in)')
```
